```python
def dedupe_columns(df):
    columns = pd.Series(df.columns)
    for dup in columns[columns.duplicated()].unique():
        dup_idx = columns[columns == dup].index.tolist()       
        for i in range(1, len(dup_idx)):
            columns[dup_idx[i]] = f"{dup}_{i}"
    df.columns = columns
    return df
```

```python
census_2012_pop_final = dedupe_columns(census_2012_pop_final)
census_2017_pop_final = dedupe_columns(census_2017_pop_final)
census_2022_pop_final = dedupe_columns(census_2022_pop_final)
```
